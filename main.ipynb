```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("Here is the COGS 108 project proposal template.")
```

```python
x = np.array([1, 3, 5])
y = np.array([8, 2, 1])

# Compute MLE estimates
a_hat = -1.75
b_hat = 8.917

# Compute regression line
x_line = np.linspace(0, 6, 100)  # More points for smooth line
y_line = a_hat * x_line + b_hat
```

```python
# Plot data
plt.scatter(x, y, color='black', label="Data points")
plt.plot(x_line, y_line, color='red', label="Fitted Line: y = -1.75x + 8.917")

# Labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression via MLE")
plt.legend()
plt.grid(True)
plt.show()
```
